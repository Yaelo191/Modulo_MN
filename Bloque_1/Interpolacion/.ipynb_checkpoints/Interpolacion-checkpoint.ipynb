## Ejercicio 1:

Considere una serie de puntos: $(x_0, y_0), (x_1, y_1), \dots, (x_n, y_n)$ y queremos encontrar el polinomio de mayor grado pase por todos ellos. Sea el conjunto de puntos:
    
$$C=\{(2,2),(3,6),(4,5),(5,5),(6,6)\}$$

- Cual sería el polinomio de mayor grado capaz de pasar por todos ellos.
 
    RESPUESTA: Polinomio de grado 4

- Cree un código que muestre el polinomio de mayor grado y grafíquelo (`puede usar Sympy`).
- Compare su resultado con np.polyfit.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

def bary_interpol_v2(puntosx, puntosy):
    x = sym.Symbol('x')
    n = puntosx.size
    pesos = np.ones_like(puntosx)
    for i in range(n):
        for j in range(n):
            if j != i:
                pesos[i] *= (puntosx[i]-puntosx[j])
    pesos = 1/pesos
    if not np.any(puntosx == x):
        numerador = np.sum(pesos*puntosy/(x-puntosx))
        denominador = np.sum(pesos/(x-puntosx))
        ecuacion = numerador/denominador
    else:
        k = np.where(x == puntosx)[0]
        ecuacion = puntosy[k[0]] 
    expresion = sym.simplify(ecuacion)
    ecuacion = sym.lambdify(x, expresion) # Convertir la función simbólica en una función numérica
    return ecuacion, expresion

C = np.array([(2, 2), (3, 6), (4, 5), (5, 5), (6, 6)])
puntosx = C[:, 0]
puntosy = C[:, 1]

lagrange, expresion = bary_interpol_v2(puntosx, puntosy)
polyfit = np.polyfit(puntosx, puntosy, 4)
polyfit = np.poly1d(polyfit)

xi = np.linspace(2, 6, 30)
plt.plot(puntosx, puntosy,'o', label = 'C puntos')
plt.plot(xi, lagrange(xi), 'r:', label = 'Lagrange{expresion}')
#plt.plot(xi, polyfit(xi), 'b-', label = 'polyfit')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

## Ejercicio 2:

Sea la función analítica:
    
$$f(x)=\frac{1}{1+x^2}$$

- Genere un juego de nodos (datos) en el intervalo $[-5, 5]$, equiespaciados y grafiquelo.
- Analizando la figura, ¿qué método de interpolación usarías polinomial (Lagrange, Newton) o de trazados (lineal, splines)?

    RESPUESTA:Un spline cubico seria ideal por la forma de la funcion en cada intervalo y por la cantidad de puntos. 
  
- Programe un método de cada clase y grafíquelo.
- ¿Qué metodo de interpolación sería el correcto a usar? Justifique su resultado graficándolo.
 
    RESPUESTA:Si seria mejor un spline cubico, usare de ejemplo el spline lineal, que aun asi se apromi mejor que un metodo polinomial. 
  

```python
f = lambda x: 1/(1+x**2)

x2 = np.linspace(-5, 5, 9)
y2 = f(x2)
plt.plot(x2, y2, 'go')
```

```python
def linear_spline_interpolation(points_x, points_y, x_interp):

    x = np.array(points_x)
    y = np.array(points_y)

    if len(x) != len(y):
        raise ValueError("points_x and points_y must have the same length")

    indices = np.searchsorted(x, x_interp)
    slopes = (y[indices] - y[indices - 1]) / (x[indices] - x[indices - 1])
    interpolated_values = y[indices - 1] + slopes * (x_interp - x[indices - 1])

    return interpolated_values

lagrange, expresion = bary_interpol_v2(x2, y2)
polyfit = np.polyfit(x2, y2, 4)
polyfit = np.poly1d(polyfit)

xi = np.linspace(-5, 5, 100)
yi = np.array([linear_spline_interpolation(x2, y2, i) for i in xi])
plt.plot(x2, y2,'o', label = 'C puntos')
plt.plot(xi, lagrange(xi), 'r:', label = 'Lagrange')
plt.plot(xi, yi, 'b', label = 'Lineal spline')
plt.plot(xi, polyfit(xi), 'g-', label = 'polyfit')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

## Ejercicio 4

Considere la función:

$$f(x)=\sin(\pi x)$$

genere un conjunto de $3$ a $5$ nodos en el intervalo $[0, \pi/2]$.

- ¿Qué método de interpolación usaría?

RESPUESTA: Interpolacion trigonometrica

- Realice la interpolación mediante el método escogido. 
- Integre la función interpolada en el intervalo $[0, \pi/2]$ usando alguno de los métodos numéricos conocidos.
- Estime el error para cada uno de los incisos anteriores.
- ¿Qué ventaja se tendría si el intervalo es $[0, 2\pi]$?
- Compare ambas implementaciones.

```python
import numpy as np
import sympy as sym

def parametros(puntosx, puntosy):
    n = len(puntosx)
    m = int((n-1)/2)
    ak = np.zeros(m+1)
    bk = np.zeros(m)  
    for i in range(m+1):
        ak[i] = (np.sum(puntosy*np.cos(i*puntosx))*2)/n
    for i in range(1, m):  
        bk[i] = (np.sum(puntosy*np.sin(i*puntosx))*2)/n
    return ak, bk
    
def inttrig_impar(puntosx, puntosy):
    x = sym.Symbol('x')
    ak, bk = parametros(puntosx, puntosy)
    n = len(ak) + len(bk)
    m = int((n-1)/2)
    val = ak[0]/2
    for i in range(1, m+1):
        val += ak[i]*sym.cos(i*x)
        val += bk[i-1]*sym.sin(i*x)
    print(ak, bk)
    return val

f3 = lambda x: np.sin(np.pi*x)
val = np.pi/2
puntosx = np.linspace(0, val, 5)
puntosy = np.array([f3(i) for i in puntosx])
print(puntosx)
print(puntosy)
f_trig = inttrig_impar(puntosx, puntosy)
print(f_trig)
```

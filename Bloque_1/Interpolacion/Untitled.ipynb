# Pueba Spline Cubico

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
    
f = lambda x: 1/(1+x**2)

x2 = np.linspace(-5, 5, 9)
y2 = f(x2)
#plt.plot(x2, y2, 'go')

def calculo_terminos(puntosx, puntosy):
    
    n = len(puntosx)
    matriz = np.eye(n, dtype=float)
    matriz *= 2.0
    matriz[0][1] = 1.0
    matriz[n-1][n-2] = 1.0
    
    Hs = [(puntosx[i+1]-puntosx[i]) for i in range(n-1)]
    Ps = [(puntosy[i+1]-puntosy[i]) for i in range(n-1)]
    Hm = [1/Hs[i] for i in range(n-1)]
    
    for i in range(1, len(Hm)):
        matriz[i][i] = 2*(Hm[i-1]+Hm[i])
        matriz[i][i-1] = Hm[i-1]
        matriz[i][i+1] = Hm[i]
    
    vector_dif = np.zeros(n, dtype=float)
    vector_dif[0] = 3*(puntosy[1]-puntosy[0])
    vector_dif[n-1] = 3*(puntosy[n-1]-puntosy[n-2])

    for i in range(1, n-1):
        frac_one = Ps[i-1]/Hs[i-1]
        frac_two = Ps[i]/Hs[i]
        vector_dif[i]=3*(frac_one + frac_two)
    D = np.array(np.linalg.solve(matriz, vector_dif))
    return Ps, D, Hs
    
def cubic_spline(puntosx, puntosy, x):
    n = len(puntosx)
    Ps, D, Hs = calculo_terminos(puntosx, puntosy)
    Si = 0
    for i in range(n-1):
        if puntosx[i] <= x <= puntosx[i+1]:
            Si = puntosy[i]
            Si += D[i]*(x-puntosx[i])
            Si += ((puntosy[i]-D[i])*(x-puntosx[i])**2)/Hs[i]
            Si += ((x-puntosx[i+1])*(D[i]+D[i+1]-2*puntosy[i])*(x-puntosx[i])**2)/(Hs[i])**2
            break
    return Si
    
print(x2[2])
print(y2[2])
print(cubic_spline(x2, y2, -2.5))

polyfit = np.polyfit(x2, y2, 9)
polyfit = np.poly1d(polyfit)

xi = np.linspace(-5, 5, 100)
yi = np.array([cubic_spline(x2, y2, i) for i in xi])

plt.plot(x2, y2,'o', label = 'C puntos')
plt.plot(xi, yi, 'r:', label = 'Cubic')
#plt.plot(xi, polyfit(xi), 'g-', label = 'polyfit')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

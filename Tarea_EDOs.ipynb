### Ejercicio 1

Hallar la solucion que pasa por $(0, 1)$ de:  
$$(1+e^x)yy'=e^x$$
$$y'=\frac{e^x}{(1+e^x)y}$$
Compare con la solucion analitica: 
$$y = ±\sqrt{2\ln{(1+e^x)}+C}$$
Para hallar el valor de la constante se sustituye con las condiciones iniciales:
$$1 = ±\sqrt{2\ln{(1+e^1)}+C}$$
Despejamos para C comenzando por elevar ambos miembors al cuadrado, lo que elimina el caso negativo de la ecuacion:
$$1 = 2\ln{(1+e)}+C$$
$$C = 1-2\ln{(1+e)}$$
Volviendo a la solucion general analitica, sustituimos C para tener la solucion particular y reorganizamos:
$$y = ±\sqrt{1+2\ln{(\frac{1+e^x}{1+e})}}$$

```python
import numpy as np 
import matplotlib.pyplot as plt
from Bloque_1.EDO.edo_euler import edo_euler, mejorado_euler
from Bloque_1.EDO.edo_rk4 import edo_rk4
```

```python
def g(x):
    return np.sqrt(1 + 2 * np.log((1 + np.exp(x)) / (1 + np.exp(1))))
def f(x,y):
    return (np.exp(x))/(1 + np.exp(x)*y)
     
a, b, n, yinit = 0, 100, 12, 1
N = 1000
x_euler, y_euler = edo_euler(f, a, b, n, yinit)
x_meuler, y_meuler = edo_euler(f, a, b, n, yinit)
x_rk4, y_rk4 = edo_rk4(f, a, b, n, yinit)
x_rk4_2, y_rk4_2 = edo_rk4(f, a, b, N, yinit)

plt.plot(x_euler, y_euler, '--', label='Euler')
plt.plot(x_meuler, y_meuler, '--', label='MEuler')
plt.plot(x_rk4, y_rk4, '--', label='RK4')
plt.plot(x_rk4_2, y_rk4_2, '--', label='RK4 1000p')
plt.plot(x_rk4, g(x_rk4), '--', label='Analitico')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de los metodos de Euler y RK4')
plt.grid(True)
```

### Ejercicio 2

Hallar la solucion que pasa por $(1, 1)$ de:  
$$\frac{dy}{dx} = \frac{x-y}{x}$$
$$y' = \frac{x-y}{x}$$
Compare con la solucion analitica: 
$$y=\frac{x}{2}-\frac{C}{x}$$
Para hallar el valor de la constante se sustituye con las condiciones iniciales:
$$1=\frac{1}{2}-\frac{C}{1}$$
Despejamos para C:
$$C=\frac{1}{2}-1$$
$$C = -\frac{1}{2}$$
Volviendo a la solucion general analitica, sustituimos C para tener la solucion particular y reorganizamos:
$$y = \frac{x^2 + 1}{2x}$$

```python
def g(x):
    return (x**2 + 1)/(2*x)
def f(x, y):
    return (x - y)/(x)
     
a, b, n, yinit = 1, 1.05, 12, 1
N = 20
x_euler, y_euler = edo_euler(f, a, b, n, yinit)
x_meuler, y_meuler = edo_euler(f, a, b, n, yinit)
x_rk4, y_rk4 = edo_rk4(f, a, b, n, yinit)
x_rk4_2, y_rk4_2 = edo_rk4(f, a, b, N, yinit)

plt.plot(x_euler, y_euler, '--', label='Euler')
plt.plot(x_meuler, y_meuler, '--', label='MEuler')
plt.plot(x_rk4, y_rk4, '--', label='RK4')
plt.plot(x_rk4_2, y_rk4_2, ':', label='RK4 20p')
plt.plot(x_euler, g(x_euler), '--', label='Analitico')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de los metodos de Euler y RK4')
plt.grid(True)
```

### Ejercicio 3

Hallar la solucion de:  
$$xy' + 4y = x^3 - x$$
$$y'= \frac{x^3 - x - 4y}{x}$$
Compare con la solucion analitica: 
$$y=\frac{x^3}{7}-\frac{x}{5}+\frac{C}{x^4}$$
Para hallar el valor de la constante se sustituye con las condiciones iniciales $y(1)=0$ que escogi por simplificar el depeje de $C$ y tener la grafica positiva a partir de ese punto:
$$0=\frac{1^3}{7}-\frac{1}{5}+\frac{C}{1^4}$$
Despejamos para C:
$$C=\frac{1}{5}-\frac{1}{7}$$
$$C = \frac{2}{35}$$
Volviendo a la solucion general analitica, sustituimos C para tener la solucion particular y reorganizamos:
$$y=\frac{x^3}{7}-\frac{x}{5}+\frac{2}{35x^4}$$

```python
def g(x):
    return (x**3)/(7)-(x)/(5)+(2)/(35*x**4)
def f(x, y):
    return (x**3 - x - 4*y)/(x)
     
a, b, n, yinit = 1, 1.5, 12, 0
N = 100
x_euler, y_euler = edo_euler(f, a, b, n, yinit)
x_meuler, y_meuler = edo_euler(f, a, b, n, yinit)
x_rk4, y_rk4 = edo_rk4(f, a, b, n, yinit)
x_rk4_2, y_rk4_2 = edo_rk4(f, a, b, N, yinit)

plt.plot(x_euler, y_euler, '--', label='Euler')
plt.plot(x_meuler, y_meuler, '--', label='MEuler')
plt.plot(x_rk4, y_rk4, '--', label='RK4')
plt.plot(x_rk4_2, y_rk4_2, '--', label='RK4 100p')
plt.plot(x_euler, g(x_euler), '--', label='Analitico')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de los metodos de Euler y RK4')
plt.grid(True)
```

### Ejercicio 4

Hallar la solucion cuando la condicion de frontera es $y(0)=1$:  
$$(x^3 + xy^2)dx + (x^2y+y^3)dy=0$$
$$y'= -\frac{x^3 + xy^2}{x^2y+y^3}$$
Compare con la solucion analitica: 
$$C = (x^2 + y^2)^2$$
Para hallar el valor de la constante se sustituye la ecuacion con las condiciones iniciales.
$$C = (0^2 + 1^2)^2$$
$$C = (1)^2$$
$$C = 1$$
Volviendo a la solucion general analitica, sustituimos C para tener la solucion particular y reorganizamos:
$$y = \sqrt{1-x^2}$$

```python
def g(x):
    return np.sqrt(1-x**2)
def f(x, y):
    return -(x**3 + x*y**2)/(y*x**2 + y**3)

a, b, n, yinit = 0, 1, 12, 1
N = 100
x_euler, y_euler = edo_euler(f, a, b, n, yinit)
x_meuler, y_meuler = edo_euler(f, a, b, n, yinit)
x_rk4, y_rk4 = edo_rk4(f, a, b, n, yinit)
x_rk4_2, y_rk4_2 = edo_rk4(f, a, b, N, yinit)

plt.plot(x_euler, y_euler, '--', label='Euler')
plt.plot(x_meuler, y_meuler, '--', label='MEuler')
plt.plot(x_rk4, y_rk4, '--', label='RK4')
plt.plot(x_rk4_2, y_rk4_2, '--', label='RK4 100p')
plt.plot(x_euler, g(x_euler), '--', label='Analitico')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de los metodos de Euler y RK4')
plt.grid(True)
```
